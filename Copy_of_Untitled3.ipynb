```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/customer_churn_dataset-training-master.csv")

df.head()
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
churn_rate = df['Churn'].mean()*100

print("Churn Rate: {:.2f}%".format(churn_rate))
```

```python
df['Churn'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(6,6)
)

plt.title("Customer Churn Distribution")
plt.ylabel('')
plt.show()
```

```python
subscription_churn = pd.crosstab(
    df['Subscription Type'],
    df['Churn']
)

subscription_churn.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Subscription Type vs Churn")
plt.ylabel("Customers")
plt.show()
```

```python
contract_churn = pd.crosstab(
    df['Contract Length'],
    df['Churn']
)

contract_churn.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title("Contract Length vs Churn")
plt.ylabel("Customers")
plt.show()
```

```python
df.groupby('Churn')['Age'].mean().plot(
    kind='bar',
    figsize=(6,4)
)

plt.title("Average Age by Churn")
plt.ylabel("Age")
plt.show()
```

```python
df.groupby('Churn')['Total Spend'].mean().plot(
    kind='bar',
    figsize=(6,4)
)

plt.title("Average Spend by Churn")
plt.ylabel("Total Spend")
plt.show()
```

```python
df.groupby('Churn')['Support Calls'].mean().plot(
    kind='bar',
    figsize=(6,4)
)

plt.title("Average Support Calls by Churn")
plt.ylabel("Support Calls")
plt.show()
```

INSIGHTS
1. A significant percentage of customers have churned.
2. Churn varies across subscription types.
3. Contract length has a strong impact on retention.
4. Customers with higher support calls tend to churn more.
5. Customer spending patterns differ between retained and churned customers.
